```python
import pandas as pd
import transformers
import numpy as np
```

```python
# Load the Davidson dataset
df = pd.read_csv("https://raw.githubusercontent.com/t-davidson/hate-speech-and-offensive-language/master/data/labeled_data.csv")
```

```python
from transformers import AlbertTokenizer, AlbertModel
tokenizer = transformers.AlbertTokenizerFast.from_pretrained('albert-base-v2')

```

```python
def preprocess_text(text):
    return tokenizer.encode_plus(
        text,
        add_special_tokens=True,
        max_length=128,
        pad_to_max_length=True,
        return_attention_mask=True,
        return_token_type_ids=False,
        truncation=True
    )
```

```python
input_ids = np.vstack(df['tweet'].apply(preprocess_text).apply(lambda x: x['input_ids']))
attention_masks = np.vstack(df['tweet'].apply(preprocess_text).apply(lambda x: x['attention_mask']))
labels = np.array(df['class'])
```

```python
!pip install scikit-learn
```

```python
import sklearn
```

```python
# Split the dataset into training and testing sets
from sklearn.model_selection import train_test_split

train_input_ids, test_input_ids, train_attention_masks, test_attention_masks, train_labels, test_labels = train_test_split(
    input_ids,
    attention_masks,
    labels,
    test_size=0.2,
    random_state=42
)

```

```python
import tensorflow as tf
print(tf.__version__)
```

```python
# Create a TensorFlow dataset from the training data
train_dataset = tf.data.Dataset.from_tensor_slices((
    {"input_ids": train_input_ids, "attention_mask": train_attention_masks},
    train_labels
))
```

```python
# Create a TensorFlow dataset from the testing data
test_dataset = tf.data.Dataset.from_tensor_slices((
    {"input_ids": test_input_ids, "attention_mask": test_attention_masks},
    test_labels
))
```

```python
model = transformers.TFAlbertForSequenceClassification.from_pretrained('albert-base-v2', num_labels=3)

# Compile the model
optimizer = tf.keras.optimizers.Adam(learning_rate=2e-5)
model.compile(optimizer=optimizer, loss=model.hf_compute_loss,metrics=['accuracy'])
# model.compile(optimizer=optimizer, loss=model.compute_loss, )
```

```python
history = model.fit(
    train_dataset.batch(32),
    epochs=8,
    validation_data=test_dataset.batch(32)
)
```

```python
# Evaluate the model
test_loss, test_accuracy = model.evaluate(test_dataset.batch(32))
print("Test loss:", test_loss)
print("Test accuracy:", test_accuracy)
```
